## Player point algorithm 

### Cano

```python

my_cano_salaries = []
my_cano_total_points = []


for (i,j,z) in zip(cano_year_list,overall_list,salary_list):
    a = np.array(i, dtype=np.float)
    b = np.array(j, dtype=np.float)
    check = a/b
    check[0] = check[0]*4
    total_points = check.sum()
    point_worth = z/8 ## 8 is the amount of points your average player gets 
    salaries = point_worth * total_points
    my_cano_total_points.append(total_points)
    my_cano_salaries.append(salaries)
    
bat_cano['points'] = my_cano_total_points
bat_cano['my_salary'] = my_cano_salaries
```
